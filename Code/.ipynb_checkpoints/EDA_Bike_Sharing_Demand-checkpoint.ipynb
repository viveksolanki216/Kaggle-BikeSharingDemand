```python

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Libraries Imported")
```

```python
def describe_data(data):
    print("-------------Size-------------")
    print(data.shape)

    print("\n\n---------Metadata Info--------")
    print(data.info())

    print("\n\n----------Sample Data---------")
    print(data.head())

    print("\n\n------------Summary-----------")
    print(data.describe().T)
```

```python
home_dir = "/home/vss/Documents/Personal/Personal Learnings - Data Science/OnlineCompetitions/Bike Sharing Demand/"
data_dir = f"{home_dir}Input/"
train = pd.read_csv(f"{data_dir}train.csv")
test = pd.read_csv(f"{data_dir}test.csv")
sampleSubm = pd.read_csv(f"{data_dir}sampleSubmission.csv")
```

```python
describe_data(train)

# 10886 Instances in the dataset
# Hourly Bike demand is given in three columns casual, registered and count
# count is just a sum of registered and casual
# registered demand is comning for registered users and casual is coming from non-reg.
# we have provided other columns too
```

```python
describe_data(test)
# Similar as train data except demand of bike rental is not provided.
```

```python
describe_data(sampleSubm)
# Hourly datetime is provided, for which we need to predict the bike demand aka "count"
```

```python
train.datetime = pd.to_datetime(train.datetime)
test.datetime = pd.to_datetime(test.datetime)


train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month

test['year'] = test['datetime'].dt.year
test['month'] = test['datetime'].dt.month
```

```python
print(train.datetime.min(), train.datetime.max())
print(test.datetime.min(), test.datetime.max())
```

```python
# Let's dig deeper in the training data

# Start with the target variable count and visualize how it looks on timeline.
fig = plt.figure(figsize=(20,5))
plt.plot(train['datetime'], train['count'], 'o', markersize='1')

# Empty gaps every month, suggests no count data is provided and we need to predict for those gaps
# Increasing trend over the time suggest increase in bike demands
# Yearly periodic, suggest in fall and summers high demand and in winter lower demand

# we have two components for the bike demand, from registered users and from casual users
fig = plt.figure(figsize=(20,5))
plt.plot(train['datetime'], train['registered'], 'ob', markersize='1')
plt.plot(train['datetime'], train['casual'], 'or', markersize='1')

# Demand from Casual users are typically lower
```

```python
# Explore other variables

# Season
# values explained : season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
print(train.groupby(['season'])['datetime'].min())
train['season'].value_counts()

# Seeing the above timeline summer and fall has high demand, while for winter and spring seems 
# to have low demands, Let's check
print(train.groupby(['season'])['count'].describe().round())
# Description shows lowest demand for Spring, while the winter has higher demands then spring

# Seems like we need to see the above summary by year
print(train.groupby(['year','season'])['count'].describe().round())
# Now if you see for 4th quqarter of 2011 and 1st quarter of 2012, you see that they have equivalent demands
# Now because of increasing trend, winter sales should be lower than the spring one.

# Now look at casual and registered user trend
print(train.groupby(['year','season'])['registered'].describe().round())
print(train.groupby(['year','season'])['casual'].describe().round())

# Seems like casual(non-registered demands is higher in fall and summers proportionally then other season)
# v/s that of registered. Tourtist could be the
```

```python
def describe_categorical_variables_vs_target(data, var_cat):
    
    print(data.groupby([var_cat])['count'].describe().round())
    print(data.groupby(['year',var_cat])['count'].describe().round())
    print(data.groupby(['year',var_cat])['registered'].describe().round())
    print(data.groupby(['year',var_cat])['casual'].describe().round())
```

```python
# Season
# weather - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
# 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
# 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
# 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
# values explained : season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
#print(train.groupby(['weather'])['datetime'].min())
print(train['weather'].value_counts())

print(train.groupby(['weather'])['count'].describe().round())
print(train.groupby(['weather'])['registered'].describe().round())
print(train.groupby(['weather'])['casual'].describe().round())

# As we see, the demand goes down as the weather goes bad.
# For the worst weather (4), as you see, there is only one observation, it has a good demand, we can't much 
# say just based on one event

print(train[ train['weather'] == 4])
```

```python
# Working day
cat_var = 'workingday'
print(train[cat_var].value_counts())

print(train.groupby([cat_var])['count'].describe().round())
print(train.groupby([cat_var])['registered'].describe().round())
print(train.groupby([cat_var])['casual'].describe().round())

# As we clealry see, "registered" demand is more on working day and "casual" demand is higher on non-working days
```

```python
# Holiday
cat_var = 'holiday'
print(train[cat_var].value_counts())

print(train.groupby([cat_var])['count'].describe().round())
print(train.groupby([cat_var])['registered'].describe().round())
print(train.groupby([cat_var])['casual'].describe().round())

# As we clealry see, opposite to what we have observed for workingday
```

```python
# For Numeric Variable
num_variables = ['temp', 'atemp', 'humidity', 'windspeed']
for num_var in num_variables:
    print(f"----------{num_var}-----------------")
    print(train[num_var].corr(train['count']).round(2))
    plt.plot(train['temp'], train['count'], 'o', markersize=1, alpha=.5)
    plt.show()
    print("Casual:")
    print(train[num_var].corr(train['casual']).round(2))
    plt.plot(train['temp'], train['casual'], 'o', markersize=1, alpha=.5)
    plt.show()
    print("Registered:")
    print(train[num_var].corr(train['registered']).round(2))
    plt.plot(train['temp'], train['registered'], 'o', markersize=1, alpha=.5)
    plt.show()
```

```python
import seaborn as sns
```

```python
# For Numeric Variable
num_variables = ['temp', 'atemp', 'humidity', 'windspeed']
quantiles = np.arange(0,1,0.05) #[0,.25,.50,.75,1]
for num_var in num_variables:
    print(f"----------{num_var}-----------------")
    
    bins = pd.cut(train[num_var], train['temp'].quantile(quantiles).tolist())
    
    #print(train.groupby([bins])['count'].describe())
    sns.boxplot(x=bins, y = train['count'])
    plt.show()
    
```

```python
# What other factor apart from the variables given will affect the demand?
# does different time of the day will effect the demand

# To create feature from existing features is called Feature Extraction

train['hour'] = train['datetime'].dt.hour

#print(train['hour'].value_counts())
# Holiday
cat_var = 'hour'

print(train.groupby([cat_var])['count'].describe().round())
print(train.groupby([cat_var])['registered'].describe().round())
print(train.groupby([cat_var])['casual'].describe().round())
```

```python
train['day'] = train['datetime'].dt.weekday

print(train['day'].value_counts())
# Holiday
cat_var = 'day'

print(train.groupby([cat_var])['count'].describe().round())
print(train.groupby([cat_var])['registered'].describe().round())
print(train.groupby([cat_var])['casual'].describe().round())
```
